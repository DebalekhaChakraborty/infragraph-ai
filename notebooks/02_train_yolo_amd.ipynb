# 02 — Train YOLOv8 (AMD / ROCm / CUDA)

Fine-tunes a YOLOv8 model on the InfraGraph AI dataset.
Works on CUDA GPUs, AMD ROCm GPUs, and CPU fallback.

```python
import torch
print('PyTorch :', torch.__version__)
print('CUDA    :', torch.cuda.is_available())
print('Device  :', torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'CPU')
```

```python
from ultralytics import YOLO
import yaml

with open('../configs/train_config.yaml') as f:
    cfg = yaml.safe_load(f)

print('Training config:')
for k, v in cfg.items():
    print(f'  {k}: {v}')
```

```python
model = YOLO(cfg['model'])

results = model.train(
    data    = '../configs/dataset_config.yaml',
    epochs  = cfg['epochs'],
    imgsz   = cfg['imgsz'],
    batch   = cfg['batch'],
    patience= cfg['patience'],
    optimizer = cfg['optimizer'],
    lr0     = cfg['lr0'],
    project = '../' + cfg['project'],
    name    = cfg['name'],
    exist_ok= cfg['exist_ok'],
    save_period = cfg['save_period'],
)
print('Training complete.')
```

```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

run_dir = Path('../outputs/infragraph_run')
for curve in ['results.png', 'confusion_matrix.png']:
    p = run_dir / curve
    if p.exists():
        plt.figure(figsize=(14, 6))
        plt.imshow(Image.open(p))
        plt.axis('off')
        plt.title(curve)
        plt.show()
```
